```python
import pandas as pd
from sklearn.linear_model import LinearRegression # Importa o modelo de regressão
import os # Para verificar o arquivo

# --- Passo 1: Carregar o DataFrame original e treinar um modelo (Exemplo) ---
file_path = "/content/Preços_de_casas.csv"

try:
    if not os.path.exists(file_path):
        raise FileNotFoundError(f"Arquivo de treinamento não encontrado: {file_path}")

    df_treino = pd.read_csv(file_path)
    if 'Id' in df_treino.columns:
        df_treino = df_treino.drop(columns = "Id")
    print("Dataset de treino carregado com sucesso!")

    # --- VERIFICAÇÃO IMPORTANTE: Exibir as colunas do DataFrame ---
    print("\nColunas disponíveis no DataFrame de treino:")
    print(df_treino.columns.tolist())
    # --- FIM DA VERIFICAÇÃO ---


    # Pre-processamento para a variável 'qualidade_da_cozinha'
    # VERIFIQUE AQUI O NOME EXATO DA SUA COLUNA DE QUALIDADE DA COZINHA!
    # Por exemplo, se na lista acima aparecer 'Qualidade_Cozinha', mude 'qualidade_da_cozinha' para 'Qualidade_Cozinha'
    coluna_qualidade_cozinha = 'qualidade_da_cozinha' # <--- AJUSTE AQUI SE O NOME DA COLUNA FOR DIFERENTE NO SEU CSV

    if coluna_qualidade_cozinha in df_treino.columns:
        df_treino = pd.get_dummies(df_treino, columns=[coluna_qualidade_cozinha], drop_first=True, dtype=int)
        print(f"Variável '{coluna_qualidade_cozinha}' transformada com sucesso em variáveis dummy.")
        print("Novas colunas após one-hot encoding:", [col for col in df_treino.columns if 'qualidade_da_cozinha' in col])
    else:
        print(f"AVISO: A coluna '{coluna_qualidade_cozinha}' não foi encontrada no dataset. Ignorando a transformação.")


    # Definir variáveis explicativas (X) e variável resposta (y)
    # As variáveis aqui devem ser EXATAMENTE as que seu modelo real usou no treino.
    # Se você não usou 'existe_segundo_andar' ou 'qualidade_da_cozinha_Excelente', remova-as daqui.
    features_para_modelo = [
        'quantidade_banheiros',
        'area_primeiro_andar',
        'existe_segundo_andar',
        'qualidade_da_cozinha_Excelente' # Esta será criada pelo get_dummies
    ]

    # Filtra as features que realmente existem no DataFrame após o pré-processamento
    available_features = [f for f in features_para_modelo if f in df_treino.columns]

    # Se a 'qualidade_da_cozinha' original não estava presente e não criou a dummy,
    # precisamos garantir que 'qualidade_da_cozinha_Excelente' não esteja em available_features
    if 'qualidade_da_cozinha_Excelente' not in df_treino.columns and 'qualidade_da_cozinha_Excelente' in available_features:
        available_features.remove('qualidade_da_cozinha_Excelente')

    # Verificar se as features necessárias para a nova casa estão disponíveis
    required_new_house_features = [
        'quantidade_banheiros', 'area_primeiro_andar',
        'existe_segundo_andar', 'qualidade_da_cozinha_Excelente'
    ]
    # Certifica-se de que estamos apenas selecionando as que foram realmente usadas no treino
    features_para_prever_nova_casa = [f for f in required_new_house_features if f in available_features]


    X_treino = df_treino[available_features]
    y_treino = df_treino['preco_de_venda']

    # Treinar o modelo de regressão linear
    modelo_regressao = LinearRegression()
    modelo_regressao.fit(X_treino, y_treino)
    print("\nModelo de regressão treinado com sucesso com as features:", available_features)

except FileNotFoundError as e:
    print(f"ERRO: {e}")
    print("Certifique-se de que o arquivo de treino 'Preços_de_casas.csv' está no local correto.")
    exit()
except KeyError as e:
    print(f"ERRO: A coluna '{e}' não foi encontrada no dataset 'df_treino'. Verifique os nomes das colunas e o pré-processamento.")
    exit()
except Exception as e:
    print(f"Ocorreu um erro inesperado ao carregar o arquivo ou treinar o modelo: {e}")
    exit()


# --- Passo 2: Preparar os dados da nova casa para previsão ---
# Crie um DataFrame com as características da nova casa
# As colunas e a ordem devem ser as mesmas usadas para treinar o modelo!
# Use os valores 0 ou 1 para 'existe_segundo_andar' e 'qualidade_da_cozinha_Excelente'
dados_nova_casa = pd.DataFrame([[1, 98, 0, 1]],
                               columns=['quantidade_banheiros', 'area_primeiro_andar',
                                        'existe_segundo_andar', 'qualidade_da_cozinha_Excelente'])

# Assegure-se de que as colunas da nova casa correspondem EXATAMENTE às features usadas no treino
# E que estão na mesma ordem!
try:
    dados_nova_casa_para_prever = dados_nova_casa[available_features]

    # --- Passo 3: Fazer a previsão ---
    preco_estimado = modelo_regressao.predict(dados_nova_casa_para_prever)

    print(f"\nCaracterísticas da nova casa para previsão: {dados_nova_casa_para_prever.iloc[0].to_dict()}")
    print(f"O preço estimado para esta casa é: R$ {preco_estimado[0]:,.2f}")

except KeyError as e:
    print(f"ERRO: A feature '{e}' necessária para a previsão da nova casa não foi usada no treinamento do modelo.")
    print("Verifique se as colunas da 'dados_nova_casa' e 'features_para_modelo' correspondem.")
except Exception as e:
    print(f"Ocorreu um erro inesperado ao preparar a nova casa ou fazer a previsão: {e}")
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, roc_curve, auc, confusion_matrix, make_scorer, f1_score
from sklearn.preprocessing import LabelEncoder
import os
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import OneHotEncoder
from imblearn.over_sampling import SMOTE
from collections import Counter

# Para gráficos
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# --- Recarregar e pré-processar os dados (garantir que X_train, X_test, y_train, y_test estão prontos) ---
file_path_churn = "/content/churn.csv"

try:
    if not os.path.exists(file_path_churn):
        raise FileNotFoundError(f"Arquivo não encontrado: {file_path_churn}")

    dados_churn = pd.read_csv(file_path_churn)

    x = dados_churn.drop(['churn', 'id_cliente'], axis=1)
    y = dados_churn['churn']

    cols_to_str = ['UF', 'Sexo', 'Cor', 'tem_cartao_credito', 'membro_ativo']
    for col in cols_to_str:
        if col in x.columns and x[col].dtype != 'object':
            x[col] = x[col].astype(str)

    colunas_categoricas_x_existentes = [col for col in ['pais', 'sexo_biologico', 'tem_cartao_credito', 'membro_ativo', 'UF', 'Cor'] if col in x.columns and x[col].dtype == 'object']

    one_hot_transformer = make_column_transformer(
        (OneHotEncoder(drop='if_binary', handle_unknown='ignore'), colunas_categoricas_x_existentes),
        remainder='passthrough',
        sparse_threshold=0
    )

    x_transformado = one_hot_transformer.fit_transform(x)
    nomes_das_novas_colunas = one_hot_transformer.get_feature_names_out(x.columns)
    df_x_transformado = pd.DataFrame(x_transformado, columns=nomes_das_novas_colunas)

    label_encoder_churn = LabelEncoder()
    y_encoded = label_encoder_churn.fit_transform(y)
    nomes_das_classes = label_encoder_churn.classes_.astype(str).tolist()

    X_train, X_test, y_train, y_test = train_test_split(df_x_transformado, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded)

    smote = SMOTE(random_state=42)
    X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)

    print("Dados carregados, pré-processados e balanceados (SMOTE) com sucesso para a otimização.")

except Exception as e:
    print(f"ERRO durante o carregamento/pré-processamento: {e}")
    exit()

# --- Etapa 3: Otimização de Hiperparâmetros para Random Forest com GridSearchCV ---
print("\n--- Iniciando Otimização de Hiperparâmetros (GridSearchCV) para Random Forest ---")

# Definir o modelo base
rf_model = RandomForestClassifier(random_state=42)

# Definir a grade de hiperparâmetros para testar
# Atenção: Este grid pode levar um tempo para rodar. Para um teste mais rápido, reduza as opções.
param_grid = {
    'n_estimators': [100, 200, 300], # Número de árvores na floresta
    'max_depth': [None, 10, 20],      # Profundidade máxima das árvores
    'min_samples_split': [2, 5],    # Número mínimo de amostras necessárias para dividir um nó interno
    'min_samples_leaf': [1, 2]      # Número mínimo de amostras necessárias para ser um nó folha
}

# Usar f1-score da classe 'churn' (classe 1) como métrica de otimização
# make_scorer permite especificar a classe para o f1_score
# Se o seu LabelEncoder mapeou 'Churn' para 1, então pos_label=1
# Se mapeou 'Churn' para 0, então pos_label=0
# Se não tiver certeza, 'f1_weighted' pode ser uma boa alternativa para considerar o peso das classes
scoring_metric = make_scorer(f1_score, pos_label=1) # Se 1 é a classe Churn, o que geralmente é o caso

# Configurar o GridSearchCV
# cv=5 significa validação cruzada de 5 folds
# verbose controla o nível de detalhe da saída
# n_jobs=-1 usa todos os núcleos disponíveis da CPU (acelera o processo)
grid_search = GridSearchCV(estimator=rf_model, param_grid=param_grid,
                           cv=5, n_jobs=-1, verbose=2, scoring=scoring_metric)

# Executar a busca na grade com os dados de TREINO balanceados
grid_search.fit(X_train_resampled, y_train_resampled)

print("\n--- Otimização de Hiperparâmetros Concluída ---")

# Exibir os melhores parâmetros encontrados
print(f"Melhores parâmetros encontrados: {grid_search.best_params_}")

# Exibir o melhor score (F1-score para Churn) obtido com esses parâmetros
print(f"Melhor F1-score (Churn) na validação cruzada: {grid_search.best_score_:.4f}")

# --- Avaliar o melhor modelo encontrado no conjunto de TESTE ---
best_rf_model = grid_search.best_estimator_
y_pred_best_rf = best_rf_model.predict(X_test)
y_proba_best_rf = best_rf_model.predict_proba(X_test)[:, 1] # Probabilidade da classe positiva

print("\n--- Avaliação do Melhor Modelo Random Forest (Otimizado) no conjunto de TESTE ---")
print(f"Acurácia (RF Otimizado): {accuracy_score(y_test, y_pred_best_rf):.4f}")
print("Relatório de Classificação (RF Otimizado):")
print(classification_report(y_test, y_pred_best_rf, target_names=nomes_das_classes))

# --- Visualizações Finais (Matriz de Confusão e Curva ROC do Modelo Otimizado) ---

# Matriz de Confusão do Modelo Otimizado
cm_best_rf = confusion_matrix(y_test, y_pred_best_rf)
plt.figure(figsize=(7, 6))
sns.heatmap(cm_best_rf, annot=True, fmt='d', cmap='Blues', cbar=False,
            xticklabels=nomes_das_classes, yticklabels=nomes_das_classes)
plt.title('Matriz de Confusão - Random Forest Otimizado')
plt.xlabel('Previsto')
plt.ylabel('Real')
plt.show()

# Curva ROC do Modelo Otimizado
fpr_best_rf, tpr_best_rf, _ = roc_curve(y_test, y_proba_best_rf)
roc_auc_best_rf = auc(fpr_best_rf, tpr_best_rf)

fig_roc_optimized = go.Figure()
fig_roc_optimized.add_trace(go.Scatter(x=fpr_best_rf, y=tpr_best_rf, mode='lines',
                                       name=f'Random Forest Otimizado (AUC = {roc_auc_best_rf:.2f})',
                                       line=dict(color='darkorange', width=2)))
fig_roc_optimized.add_trace(go.Scatter(x=[0, 1], y=[0, 1], mode='lines', name='Aleatório',
                                       line=dict(dash='dash', color='gray')))

fig_roc_optimized.update_layout(
    title='Curva ROC - Random Forest Otimizado',
    xaxis_title='Taxa de Falso Positivo (False Positive Rate)',
    yaxis_title='Taxa de Verdadeiro Positivo (True Positive Rate)',
    hovermode='x unified',
    template='plotly_white'
)
fig_roc_optimized.show()

print("\nOtimização e avaliação final do modelo Random Forest concluídas.")
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_curve, auc, make_scorer, f1_score
from sklearn.preprocessing import LabelEncoder
import os
from imblearn.over_sampling import SMOTE
from collections import Counter
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import OneHotEncoder

# Para gráficos
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go

# --- Recarregar e Pré-processar Dados ---
# (Este bloco é mantido para garantir que o script seja autocontido e reprodutível)
file_path_churn = "/content/churn.csv"

try:
    if not os.path.exists(file_path_churn):
        raise FileNotFoundError(f"Arquivo não encontrado: {file_path_churn}")

    dados_churn = pd.read_csv(file_path_churn)

    x = dados_churn.drop(['churn', 'id_cliente'], axis=1)
    y = dados_churn['churn']

    cols_to_str = ['UF', 'Sexo', 'Cor', 'tem_cartao_credito', 'membro_ativo']
    for col in cols_to_str:
        if col in x.columns and x[col].dtype != 'object':
            x[col] = x[col].astype(str)

    colunas_categoricas_x_existentes = [col for col in ['pais', 'sexo_biologico', 'tem_cartao_credito', 'membro_ativo', 'UF', 'Cor'] if col in x.columns and x[col].dtype == 'object']

    one_hot_transformer = make_column_transformer(
        (OneHotEncoder(drop='if_binary', handle_unknown='ignore'), colunas_categoricas_x_existentes),
        remainder='passthrough',
        sparse_threshold=0
    )

    x_transformado = one_hot_transformer.fit_transform(x)
    nomes_das_novas_colunas = one_hot_transformer.get_feature_names_out(x.columns)
    df_x_transformado = pd.DataFrame(x_transformado, columns=nomes_das_novas_colunas)

    label_encoder_churn = LabelEncoder()
    y_encoded = label_encoder_churn.fit_transform(y)
    nomes_das_classes = label_encoder_churn.classes_.astype(str).tolist()

    X_train, X_test, y_train, y_test = train_test_split(df_x_transformado, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded)

    smote = SMOTE(random_state=42)
    X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)

    print("Dados carregados, pré-processados e balanceados (SMOTE) com sucesso.")

except FileNotFoundError as e:
    print(f"ERRO: {e}")
    print("Certifique-se de que o arquivo 'churn.csv' está no local correto.")
    exit()
except Exception as e:
    print(f"Ocorreu um erro inesperado durante o pré-processamento: {e}")
    exit()

# --- Normalização dos Dados (RE-EXECUTAR para garantir que X_train_scaled e X_test_scaled estão prontos) ---
print("\n--- Normalizando os dados para o Modelo KNN (Re-executando) ---")

original_numeric_cols = x.select_dtypes(include=['number']).columns.tolist()
if 'id_cliente' in original_numeric_cols: original_numeric_cols.remove('id_cliente')
if 'churn' in original_numeric_cols: original_numeric_cols.remove('churn')

columns_to_scale = [col for col in original_numeric_cols if col in df_x_transformado.columns]

if not columns_to_scale:
    print("Atenção: Nenhuma coluna numérica clara encontrada para padronização. Tentando identificar colunas numéricas restantes.")
    temp_cols_to_scale = df_x_transformado.select_dtypes(include=['float64', 'int64']).columns.tolist()
    binary_cols = [col for col in temp_cols_to_scale if df_x_transformado[col].nunique() == 2 and df_x_transformado[col].isin([0, 1]).all()]
    columns_to_scale = [col for col in temp_cols_to_scale if col not in binary_cols]


print(f"Colunas numéricas selecionadas para StandardScaler: {columns_to_scale}")

scaler = StandardScaler()

X_train_scaled = X_train_resampled.copy()
if columns_to_scale:
    X_train_scaled[columns_to_scale] = scaler.fit_transform(X_train_resampled[columns_to_scale])

X_test_scaled = X_test.copy()
if columns_to_scale:
    X_test_scaled[columns_to_scale] = scaler.transform(X_test[columns_to_scale])

print("Dados normalizados com sucesso para otimização do KNN.")

# --- Otimização de Hiperparâmetros (GridSearchCV) para KNN ---
print("\n--- Iniciando Otimização de Hiperparâmetros (GridSearchCV) para KNN ---")

# Definir o modelo base do KNN
knn_model = KNeighborsClassifier()

# Definir a grade de hiperparâmetros para testar
# Valores de k comuns: 1, 3, 5, 7, ..., até um certo limite
param_grid_knn = {
    'n_neighbors': [3, 5, 7, 9, 11, 13, 15], # Testar diferentes valores de k
    'weights': ['uniform', 'distance'],      # Como os vizinhos votam
    'metric': ['euclidean', 'minkowski'] # Métrica de distância. Minkowski com p=2 é euclidean
}

# Usar f1-score da classe 'churn' (classe 1) como métrica de otimização
scoring_metric_knn = make_scorer(f1_score, pos_label=1)

# Configurar o GridSearchCV para KNN
grid_search_knn = GridSearchCV(estimator=knn_model, param_grid=param_grid_knn,
                               cv=5, n_jobs=-1, verbose=2, scoring=scoring_metric_knn)

# Executar a busca na grade com os dados de TREINO balanceados e NORMALIZADOS
grid_search_knn.fit(X_train_scaled, y_train_resampled)

print("\n--- Otimização de Hiperparâmetros KNN Concluída ---")

# Exibir os melhores parâmetros encontrados
print(f"Melhores parâmetros encontrados para KNN: {grid_search_knn.best_params_}")

# Exibir o melhor score (F1-score para Churn) obtido com esses parâmetros
print(f"Melhor F1-score (Churn) na validação cruzada para KNN: {grid_search_knn.best_score_:.4f}")

# --- Avaliar o melhor modelo KNN encontrado no conjunto de TESTE ---
best_knn_model = grid_search_knn.best_estimator_
y_pred_best_knn = best_knn_model.predict(X_test_scaled) # Usar X_test_scaled!
y_proba_best_knn = best_knn_model.predict_proba(X_test_scaled)[:, 1] # Probabilidade da classe positiva

print("\n--- Avaliação do Melhor Modelo KNN (Otimizado) no conjunto de TESTE ---")
print(f"Acurácia (KNN Otimizado): {accuracy_score(y_test, y_pred_best_knn):.4f}")
print("Relatório de Classificação (KNN Otimizado):")
print(classification_report(y_test, y_pred_best_knn, target_names=nomes_das_classes))

# --- Visualizações Finais para o Modelo KNN Otimizado ---

# Matriz de Confusão do Modelo KNN Otimizado
cm_best_knn = confusion_matrix(y_test, y_pred_best_knn)
plt.figure(figsize=(8, 6))
sns.heatmap(cm_best_knn, annot=True, fmt='d', cmap='Purples', cbar=False, # Nova cor para diferenciar
            xticklabels=nomes_das_classes, yticklabels=nomes_das_classes,
            linewidths=.5, linecolor='black')
plt.title('Matriz de Confusão - KNN Otimizado', fontsize=16)
plt.xlabel('Previsto', fontsize=12)
plt.ylabel('Real', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10, rotation=0)
plt.show()

# Curva ROC do Modelo KNN Otimizado
fpr_best_knn, tpr_best_knn, _ = roc_curve(y_test, y_proba_best_knn)
roc_auc_best_knn = auc(fpr_best_knn, tpr_best_knn)

fig_roc_optimized_knn = go.Figure()
fig_roc_optimized_knn.add_trace(go.Scatter(x=fpr_best_knn, y=tpr_best_knn, mode='lines',
                                           name=f'KNN Otimizado (AUC = {roc_auc_best_knn:.2f})',
                                           line=dict(color='darkmagenta', width=3)))
fig_roc_optimized_knn.add_trace(go.Scatter(x=[0, 1], y=[0, 1], mode='lines', name='Classificador Aleatório',
                                           line=dict(dash='dash', color='gray', width=2)))

fig_roc_optimized_knn.update_layout(
    title='Curva ROC - KNN Otimizado',
    xaxis_title='Taxa de Falso Positivo (False Positive Rate)',
    yaxis_title='Taxa de Verdadeiro Positivo (True Positive Rate)',
    hovermode='x unified',
    template='plotly_white',
    font=dict(family="Arial", size=12, color="#7f7f7f"),
    xaxis=dict(showgrid=True, gridcolor='lightgray'),
    yaxis=dict(showgrid=True, gridcolor='lightgray'),
    legend=dict(x=0.01, y=0.99, bordercolor="Black", borderwidth=1)
)
fig_roc_optimized_knn.show()

print("\nOtimização e avaliação final do modelo KNN concluídas.")
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, LabelEncoder, OneHotEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.dummy import DummyClassifier
from sklearn.metrics import accuracy_score, classification_report, f1_score, make_scorer
from sklearn.compose import make_column_transformer
import os
import plotly.graph_objects as go
import plotly.express as px
from imblearn.over_sampling import SMOTE


# --- Recarregar e Pré-processar Dados ---
file_path_churn = "/content/churn.csv"

# Initialize variables to ensure they are always defined, even if an error occurs early
# This prevents NameError in later blocks if preprocessing fails
X_train_resampled = None
y_train_resampled = None
X_train_scaled = None
X_test_scaled = None
df_x_transformado = None # Ensure this is also defined for later use
nomes_das_classes = ['0', '1'] # Default in case LabelEncoder fails or is skipped

try:
    if not os.path.exists(file_path_churn):
        raise FileNotFoundError(f"Arquivo não encontrado: {file_path_churn}")

    dados_churn = pd.read_csv(file_path_churn)

    x = dados_churn.drop(['churn', 'id_cliente'], axis=1)
    y = dados_churn['churn']

    cols_to_str = ['UF', 'Sexo', 'Cor', 'tem_cartao_credito', 'membro_ativo']
    for col in cols_to_str:
        if col in x.columns and x[col].dtype != 'object':
            x[col] = x[col].astype(str)

    colunas_categoricas_x_existentes = [col for col in ['pais', 'sexo_biologico', 'tem_cartao_credito', 'membro_ativo', 'UF', 'Cor'] if col in x.columns and x[col].dtype == 'object']

    one_hot_transformer = make_column_transformer(
        (OneHotEncoder(drop='if_binary', handle_unknown='ignore'), colunas_categoricas_x_existentes),
        remainder='passthrough',
        sparse_threshold=0
    )

    x_transformado = one_hot_transformer.fit_transform(x)
    nomes_das_novas_colunas = one_hot_transformer.get_feature_names_out(x.columns)
    df_x_transformado = pd.DataFrame(x_transformado, columns=nomes_das_novas_colunas)

    label_encoder_churn = LabelEncoder()
    y_encoded = label_encoder_churn.fit_transform(y)
    nomes_das_classes = label_encoder_churn.classes_.astype(str).tolist() # Update names after encoding

    X_train, X_test, y_train, y_test = train_test_split(df_x_transformado, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded)

    smote = SMOTE(random_state=42)
    # Corrected usage: SMOTE should be applied to X_train and y_train
    X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)

    print("Dados carregados, pré-processados e balanceados (SMOTE) com sucesso.")

except Exception as e:
    print(f"ERRO durante o carregamento/pré-processamento: {e}")
    # Exit or handle the error gracefully if data loading/preprocessing fails
    exit()

# --- Normalização dos Dados (Para KNN e outros modelos que precisarem) ---
# Ensure df_x_transformado exists from the try block
if df_x_transformado is not None:
    original_numeric_cols = x.select_dtypes(include=['number']).columns.tolist()
    if 'id_cliente' in original_numeric_cols: original_numeric_cols.remove('id_cliente')
    if 'churn' in original_numeric_cols: original_numeric_cols.remove('churn')

    columns_to_scale = [col for col in original_numeric_cols if col in df_x_transformado.columns]

    if not columns_to_scale:
        temp_cols_to_scale = df_x_transformado.select_dtypes(include=['float64', 'int64']).columns.tolist()
        binary_cols = [col for col in temp_cols_to_scale if df_x_transformado[col].nunique() == 2 and df_x_transformado[col].isin([0, 1]).all()]
        columns_to_scale = [col for col in temp_cols_to_scale if col not in binary_cols]

    scaler = StandardScaler()

    # Ensure X_train_resampled is defined before attempting to copy
    if X_train_resampled is not None:
        X_train_scaled = X_train_resampled.copy()
        if columns_to_scale:
            X_train_scaled[columns_to_scale] = scaler.fit_transform(X_train_resampled[columns_to_scale])

        X_test_scaled = X_test.copy()
        if columns_to_scale:
            X_test_scaled[columns_to_scale] = scaler.transform(X_test[columns_to_scale])
    else:
        print("Aviso: X_train_resampled não definido. Normalização de dados não pode ser realizada.")
        # Fallback for scaled data if SMOTE failed
        X_train_scaled = X_train.copy() # Use unresampled data as fallback
        X_test_scaled = X_test.copy()
        if columns_to_scale: # Attempt to scale if possible
             X_train_scaled[columns_to_scale] = scaler.fit_transform(X_train[columns_to_scale])
             X_test_scaled[columns_to_scale] = scaler.transform(X_test[columns_to_scale])

else:
    print("Erro: df_x_transformado não foi criado. Não é possível continuar com a normalização.")
    exit()


# --- Re-treinamento dos Modelos para Coletar Métricas ---
print("\n--- Re-treinando modelos para coletar métricas de comparação ---")

# Dummy Classifier
dummy_model = DummyClassifier(strategy="most_frequent", random_state=42)
dummy_model.fit(X_train_resampled, y_train_resampled)
# For Dummy, Logistic, Decision Tree, we predict on X_test (original, not scaled)
y_pred_dummy = dummy_model.predict(X_test)

# Logistic Regression
logreg_model = LogisticRegression(random_state=42, solver='liblinear', max_iter=1000)
logreg_model.fit(X_train_resampled, y_train_resampled)
y_pred_logreg = logreg_model.predict(X_test)

# Decision Tree Classifier (max_depth=4)
dt_model = DecisionTreeClassifier(max_depth=4, random_state=42)
dt_model.fit(X_train_resampled, y_train_resampled)
y_pred_dt = dt_model.predict(X_test)

# Random Forest Otimizado
rf_model_base = RandomForestClassifier(random_state=42)
param_grid_rf = {
    'n_estimators': [100],
    'max_depth': [10],
}
grid_search_rf = GridSearchCV(estimator=rf_model_base, param_grid=param_grid_rf, cv=3, n_jobs=-1, verbose=0, scoring=make_scorer(f1_score, pos_label=1))
grid_search_rf.fit(X_train_resampled, y_train_resampled)
best_rf_model = grid_search_rf.best_estimator_
y_pred_rf = best_rf_model.predict(X_test)


# KNN Otimizado
knn_model_base = KNeighborsClassifier()
param_grid_knn = { # Corrected variable name from param_knn to param_grid_knn
    'n_neighbors': [3],
    'weights': ['distance'],
    'metric': ['euclidean']
}
grid_search_knn = GridSearchCV(estimator=knn_model_base, param_grid=param_grid_knn, cv=3, n_jobs=-1, verbose=0, scoring=make_scorer(f1_score, pos_label=1))
grid_search_knn.fit(X_train_scaled, y_train_resampled)
best_knn_model = grid_search_knn.best_estimator_
y_pred_knn = best_knn_model.predict(X_test_scaled) # KNN uses scaled data for inference


print("\nMétricas sendo coletadas para comparação...")

# --- Coletar Métricas para o Gráfico ---
model_names = []
accuracies = []
f1_churn = []
recall_churn = []
precision_churn = []

# Create a dictionary of models, ready to predict on their respective test sets
# This ensures that we use the correct X_test (scaled or not) for each model type
models_for_metrics = {
    "Dummy Classifier": {"model": dummy_model, "X_test": X_test},
    "Logistic Regression": {"model": logreg_model, "X_test": X_test},
    "Decision Tree (max_depth=4)": {"model": dt_model, "X_test": X_test},
    "Random Forest (Otimizado)": {"model": best_rf_model, "X_test": X_test},
    "KNN (Otimizado)": {"model": best_knn_model, "X_test": X_test_scaled} # KNN uses scaled test data
}

for name, model_info in models_for_metrics.items():
    model = model_info["model"]
    X_test_for_prediction = model_info["X_test"]

    y_pred = model.predict(X_test_for_prediction)

    model_names.append(name)
    accuracies.append(accuracy_score(y_test, y_pred))

    report = classification_report(y_test, y_pred, target_names=nomes_das_classes, output_dict=True)

    # Handle cases where a class might have no true samples or predictions (e.g., DummyClassifier's '1' class)
    f1_churn.append(report['1'].get('f1-score', 0.0))
    recall_churn.append(report['1'].get('recall', 0.0))
    precision_churn.append(report['1'].get('precision', 0.0))

# Criar DataFrame para Plotly
df_comparison = pd.DataFrame({
    'Modelo': model_names,
    'Acurácia': accuracies,
    'F1-score (Churn)': f1_churn,
    'Recall (Churn)': recall_churn,
    'Precision (Churn)': precision_churn
})

# --- Criar o Gráfico de Barras Comparativo (Plotly) ---
fig = go.Figure()

# Adicionar barras para cada métrica
fig.add_trace(go.Bar(
    x=df_comparison['Modelo'],
    y=df_comparison['Acurácia'],
    name='Acurácia',
    marker_color='royalblue',
    text=df_comparison['Acurácia'].apply(lambda x: f'{x:.2f}'),
    textposition='outside'
))

fig.add_trace(go.Bar(
    x=df_comparison['Modelo'],
    y=df_comparison['F1-score (Churn)'],
    name='F1-score (Churn)',
    marker_color='indianred',
    text=df_comparison['F1-score (Churn)'].apply(lambda x: f'{x:.2f}'),
    textposition='outside'
))

fig.add_trace(go.Bar(
    x=df_comparison['Modelo'],
    y=df_comparison['Recall (Churn)'],
    name='Recall (Churn)',
    marker_color='lightseagreen',
    text=df_comparison['Recall (Churn)'].apply(lambda x: f'{x:.2f}'),
    textposition='outside'
))

fig.add_trace(go.Bar(
    x=df_comparison['Modelo'],
    y=df_comparison['Precision (Churn)'],
    name='Precision (Churn)',
    marker_color='mediumpurple',
    text=df_comparison['Precision (Churn)'].apply(lambda x: f'{x:.2f}'),
    textposition='outside'
))

fig.update_layout(
    title='Comparação de Desempenho dos Modelos de Previsão de Churn',
    yaxis_title='Valor da Métrica',
    xaxis_title='Modelo',
    barmode='group',
    height=600,
    hovermode="x unified",
    template='plotly_white',
    font=dict(family="Arial", size=12, color="#7f7f7f"),
    legend=dict(orientation="h", yanchor="bottom", y=1.02, xanchor="right", x=1)
)

fig.update_yaxes(range=[0, 1.05])

fig.show()

print("\nGráfico comparativo de desempenho dos modelos gerado com sucesso!")
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, LabelEncoder, OneHotEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.dummy import DummyClassifier
from sklearn.metrics import f1_score, make_scorer
from sklearn.compose import make_column_transformer
import os
import pickle

# --- Recarregar e Pré-processar Dados ---
file_path_churn = "/content/churn.csv"

# Inicializa variáveis para garantir que estejam sempre definidas
X_train_resampled = None
y_train_resampled = None
X_train_scaled = None
X_test_scaled = None
df_x_transformado = None
nomes_das_classes = ['0', '1'] # Valor padrão em caso de erro

try:
    if not os.path.exists(file_path_churn):
        raise FileNotFoundError(f"Arquivo não encontrado: {file_path_churn}")

    dados_churn = pd.read_csv(file_path_churn)

    x = dados_churn.drop(['churn', 'id_cliente'], axis=1)
    y = dados_churn['churn']

    cols_to_str = ['UF', 'Sexo', 'Cor', 'tem_cartao_credito', 'membro_ativo']
    for col in cols_to_str:
        if col in x.columns and x[col].dtype != 'object':
            x[col] = x[col].astype(str)

    colunas_categoricas_x_existentes = [col for col in ['pais', 'sexo_biologico', 'tem_cartao_credito', 'membro_ativo', 'UF', 'Cor'] if col in x.columns and x[col].dtype == 'object']

    one_hot_transformer = make_column_transformer(
        (OneHotEncoder(drop='if_binary', handle_unknown='ignore'), colunas_categoricas_x_existentes),
        remainder='passthrough',
        sparse_threshold=0
    )

    x_transformado = one_hot_transformer.fit_transform(x)
    nomes_das_novas_colunas = one_hot_transformer.get_feature_names_out(x.columns)
    df_x_transformado = pd.DataFrame(x_transformado, columns=nomes_das_novas_colunas)

    label_encoder_churn = LabelEncoder()
    y_encoded = label_encoder_churn.fit_transform(y)
    nomes_das_classes = label_encoder_churn.classes_.astype(str).tolist()

    X_train, X_test, y_train, y_test = train_test_split(df_x_transformado, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded)

    smote = SMOTE(random_state=42)
    X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)

    print("Dados carregados, pré-processados e balanceados (SMOTE) com sucesso.")

except Exception as e:
    print(f"ERRO durante o carregamento/pré-processamento: {e}")
    exit()

# --- Normalização dos Dados (Para KNN e outros modelos que precisarem) ---
# Garante que df_x_transformado foi criado com sucesso
if df_x_transformado is not None:
    original_numeric_cols = x.select_dtypes(include=['number']).columns.tolist()
    if 'id_cliente' in original_numeric_cols: original_numeric_cols.remove('id_cliente')
    if 'churn' in original_numeric_cols: original_numeric_cols.remove('churn')

    columns_to_scale = [col for col in original_numeric_cols if col in df_x_transformado.columns]

    if not columns_to_scale:
        temp_cols_to_scale = df_x_transformado.select_dtypes(include=['float64', 'int64']).columns.tolist()
        binary_cols = [col for col in temp_cols_to_scale if df_x_transformado[col].nunique() == 2 and df_x_transformado[col].isin([0, 1]).all()]
        columns_to_scale = [col for col in temp_cols_to_scale if col not in binary_cols]

    scaler = StandardScaler()

    # Garante que X_train_resampled foi criado com sucesso antes de escalar
    if X_train_resampled is not None:
        X_train_scaled = X_train_resampled.copy()
        if columns_to_scale:
            X_train_scaled[columns_to_scale] = scaler.fit_transform(X_train_resampled[columns_to_scale])

        X_test_scaled = X_test.copy()
        if columns_to_scale:
            X_test_scaled[columns_to_scale] = scaler.transform(X_test[columns_to_scale])
    else:
        print("Aviso: X_train_resampled não definido. Normalização de dados para o KNN pode ser impactada.")
        # Fallback: se o SMOTE falhou, tenta usar os dados não-resampled para escalar
        X_train_scaled = X_train.copy()
        X_test_scaled = X_test.copy()
        if columns_to_scale:
             X_train_scaled[columns_to_scale] = scaler.fit_transform(X_train[columns_to_scale])
             X_test_scaled[columns_to_scale] = scaler.transform(X_test[columns_to_scale])

else:
    print("Erro: df_x_transformado não foi criado. Não é possível continuar com a normalização ou treinamento de modelos.")
    exit()


# --- Treinamento/Otimização e Salvamento dos Modelos ---
print("\n--- Treinando/Otimizando e Salvando Modelos ---")

# Criar diretório para salvar os modelos, se não existir
output_dir = "modelos_churn_salvos"
os.makedirs(output_dir, exist_ok=True)

# 1. Salvar OneHotEncoder
with open(os.path.join(output_dir, 'one_hot_encoder.pkl'), 'wb') as f:
    pickle.dump(one_hot_transformer, f)
print(f"OneHotEncoder salvo em '{os.path.join(output_dir, 'one_hot_encoder.pkl')}'")

# 2. Salvar StandardScaler (pois KNN o usa e outros modelos numéricos podem ser adicionados)
with open(os.path.join(output_dir, 'standard_scaler.pkl'), 'wb') as f:
    pickle.dump(scaler, f)
print(f"StandardScaler salvo em '{os.path.join(output_dir, 'standard_scaler.pkl')}'")


# 3. Dummy Classifier
print("Treinando e salvando Dummy Classifier...")
dummy_model = DummyClassifier(strategy="most_frequent", random_state=42)
dummy_model.fit(X_train_resampled, y_train_resampled)
with open(os.path.join(output_dir, 'dummy_classifier_model.pkl'), 'wb') as f:
    pickle.dump(dummy_model, f)
print(f"Dummy Classifier salvo em '{os.path.join(output_dir, 'dummy_classifier_model.pkl')}'")

# 4. Logistic Regression
print("Treinando e salvando Logistic Regression...")
logreg_model = LogisticRegression(random_state=42, solver='liblinear', max_iter=1000)
logreg_model.fit(X_train_resampled, y_train_resampled)
with open(os.path.join(output_dir, 'logistic_regression_model.pkl'), 'wb') as f:
    pickle.dump(logreg_model, f)
print(f"Logistic Regression salvo em '{os.path.join(output_dir, 'logistic_regression_model.pkl')}'")

# 5. Decision Tree Classifier (max_depth=4)
print("Treinando e salvando Decision Tree Classifier...")
dt_model = DecisionTreeClassifier(max_depth=4, random_state=42)
dt_model.fit(X_train_resampled, y_train_resampled)
with open(os.path.join(output_dir, 'decision_tree_model.pkl'), 'wb') as f:
    pickle.dump(dt_model, f)
print(f"Decision Tree Classifier salvo em '{os.path.join(output_dir, 'decision_tree_model.pkl')}'")

# 6. Random Forest Otimizado
print("Otimizando e salvando Random Forest Classifier...")
rf_model_base = RandomForestClassifier(random_state=42)
param_grid_rf = {
    'n_estimators': [100], # Valores reduzidos para agilizar o salvamento, use os da sua melhor otimização
    'max_depth': [10],
}
grid_search_rf = GridSearchCV(estimator=rf_model_base, param_grid=param_grid_rf, cv=3, n_jobs=-1, verbose=0, scoring=make_scorer(f1_score, pos_label=1))
grid_search_rf.fit(X_train_resampled, y_train_resampled)
best_rf_model = grid_search_rf.best_estimator_
with open(os.path.join(output_dir, 'random_forest_optimized_model.pkl'), 'wb') as f:
    pickle.dump(best_rf_model, f)
print(f"Random Forest Otimizado salvo em '{os.path.join(output_dir, 'random_forest_optimized_model.pkl')}'")

# 7. KNN Otimizado
print("Otimizando e salvando KNN Classifier...")
knn_model_base = KNeighborsClassifier()
param_grid_knn = {
    'n_neighbors': [3], # Valores reduzidos para agilizar o salvamento, use os da sua melhor otimização
    'weights': ['distance'],
    'metric': ['euclidean']
}
grid_search_knn = GridSearchCV(estimator=knn_model_base, param_grid=param_grid_knn, cv=3, n_jobs=-1, verbose=0, scoring=make_scorer(f1_score, pos_label=1))
grid_search_knn.fit(X_train_scaled, y_train_resampled)
best_knn_model = grid_search_knn.best_estimator_
with open(os.path.join(output_dir, 'knn_optimized_model.pkl'), 'wb') as f:
    pickle.dump(best_knn_model, f)
print(f"KNN Otimizado salvo em '{os.path.join(output_dir, 'knn_optimized_model.pkl')}'")

print("\nTodos os modelos e transformadores foram salvos com sucesso na pasta 'modelos_churn_salvos'!")
```

Implicações para o Negócio

A escolha do "melhor" modelo sempre dependerá dos objetivos de negócio.

    Se o custo de um falso positivo (prever churn para quem não vai) for alto (ex: oferecer descontos caros desnecessariamente), a alta Precision do Random Forest seria ideal.

    Se o custo de um falso negativo (não prever churn para quem vai) for alto (ex: perder um cliente valioso), um modelo com alto Recall (como o KNN) poderia ser preferível, mesmo com mais falsos positivos.

Considerando o equilíbrio e a robustez geral demonstrados, o Random Forest Otimizado é a escolha mais recomendada para a previsão de churn neste cenário. Sua capacidade de identificar churners com alta confiabilidade e sua performance geral o tornam um ativo valioso para o negócio.
